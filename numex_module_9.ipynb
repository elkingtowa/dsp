Digital Communication Systems
=============================

We will remain here in a numerical world, even if, in several
applications, modulation is used for analog transmission (think of
radio signals, for instance).

Remember that in modulation there are two main players: The
modulating signal (message to be transmitted which contain the
information) and the carrier signal (which will "carry" the
information).

Good old AM radios!
-------------------

Amplitude modulation is obtained by changing the amplitude of the
carrier signal according to the amplitude of the modulating signal.

#### A toy example

Let us start with a simple example: We take as a modulating signal,
that is, the message we want to transmit,&nbsp; a simple A note,
that is a sinusoid with frequency 440Hz, and we modulated it with a
carrier sinusoid of 5KHz (recall that the frequency of the carrier
must be bigger than the frequency of the modulating signal).<br>

In Python, we can create the signals as follows

```python
numberSamples = 200000
samplingFrequ = 50000 # We assume here a sampling frequency of 50kHz

import numpy as np
timeStamps = np.arange(1,numberSamples + 1, dtype=float)/float(samplingFrequ)

ampMsg=1 # amplitude of the modulating signal (sinusoid)
ampCar=1  # amplitude of the carrier signal (sinusoid)
freqMsg=440  # frequency of the modulating sinusoid: a simple A note
freqCar=5000  # frequency of the carrier: 5KHz, with freqCar>freqMsg
modIndex=0.5  # modulation index

from scipy import constants as c

signalMsg = ampMsg*np.sin(2*c.pi*freqMsg*timeStamps)  # Message Signal

signalCar = ampCar*np.sin(2*c.pi*freqCar*timeStamps)   # Carrier Signal

signalModulated = (1 + modIndex * signalMsg)*signalCar # Amplitude Modulated signal (AM Signal)
```

Let us plot them to see how they look like

```python
%pylab inline

windowSamples = np.arange(500)  # select a visualization window composed of the first 500 samples

subplot2grid((3,3), (0,0), colspan=3)
plot(windowSamples,signalMsg[0:500])
ylabel('Message Signal')
subplot2grid((3,3), (1,0), colspan=3)
plot(windowSamples,signalCar[0:500])
ylabel('Carrier Signal')
subplot2grid((3,3), (2,0), colspan=3)
plot(windowSamples,signalModulated[0:500])
xlabel('Samples')
ylabel('AM Signal')
show()
```

The multiplication of two signal corresponds, in the frequency domain, to the convolution between the two spectra. Given that the spectrum of the carrier signal is (always) a two line spectrum,  the convolution of the spectrum of the modulating signal with a two line spectrum can be interpreted as the shifting of the modulating signal spectrum around the two lines. Let's check this out!

```python
from scipy import fftpack as f

SignalMsg = f.fft(signalMsg) # compute the DFT/DFS of the modulating signal
SignalCar = f.fft(signalCar)  # compute the DFT/DFS of the carrier signal
SignalModulated = f.fft(signalModulated) # compute the DFT/DFS of the modulated signal

normFrequ = np.arange(1,numberSamples+1,dtype = float)/float(numberSamples)
```

These are the plots of the magnitude of each FFT

```python
subplot2grid((3,3), (0,0), colspan=3)
plot(normFrequ,abs(SignalMsg))
ylabel('|Message Signal|')
subplot2grid((3,3), (1,0), colspan=3)
plot(normFrequ,abs(SignalCar))
ylabel('|Carrier Signal|')
subplot2grid((3,3), (2,0), colspan=3)
plot(normFrequ,abs(SignalModulated))
xlabel('Samples')
ylabel('|AM Signal|')
show()
```

In accordance with the theory, we see the spectrum of the modulating
signal has been "shifted" around the two spectral lines of the
carrier signal spectrum.

#### Modulating the jingle

Let us now modulate the jingle, and more precisely a low passed
version of it that can be downloaded [here]("https://spark-public.s3.amazonaws.com/dsp/num_examples/Module%209/code/jingleLP.mat).
The maximum frequency of the low passed jingle is 2.2KHz.

We start by loading the low passed jingle an by normalizing it so
that its max amplitude (in absolute value) is one.

```python
# Loading the jingle
import scipy.io as sio
mat_contents = sio.loadmat('jingleLP.mat')
Fs = mat_contents['Fs']
jingleLP = mat_contents['jingleLP']
jingleLP = jingleLP.reshape(jingleLP.size,)

numberSamples = len(jingleLP)
samplingFreq = Fs # we use as sampling frequency the one of the jingle

maxValue = max(abs(jingleLP))  # Normalize the jingle signal
ampMsg = 100 # Amplitude of modulating signal
signalMsg = ampMsg * jingleLP[0:timeStamps.size]/float(maxValue) # Modulating signal
```

Notice that, given the low dynamic of the signal, we have set the amplitude of the modulating signal to 100 for visualization pourpose.

We now create the carrier and the modulated signals

```python
ampCar = 1   # amplitude of carrier signal
freqCar = 8000   # carrier frequency of 8KHz, with freqCar>maximum frequency of modulating signal (=2.2KHz)
modIndex = 0.5   #modulation index

from scipy import constants as c
signalCar = ampCar * np.sin(2*c.pi*freqCar*timeStamps)  # Carrier Signal

signalModulated = (1 + modIndex*signalMsg)*signalCar  # Amplitude Modulated Signal
```

and plot them over the interval where there is a significant change in the signal amplitude (for visualization purpose), that is, between 63700 to 64400.

```python
windowSample = np.arange(63700,64400)

subplot2grid((3,3), (0,0), colspan=3)
plot(windowSample,signalMsg[63700:64400])
ylabel('Message Signal')
subplot2grid((3,3), (1,0), colspan=3)
plot(windowSample,signalCar[63700:64400])
ylabel('Carrier Signal')
subplot2grid((3,3), (2,0), colspan=3)
plot(windowSample,signalModulated[63700:64400])
xlabel('Samples')
ylabel('AM Signal')
show()
```

Let's see what happens in the frequency domain

```python
from scipy import fftpack as f
SignalMsg = f.fft(signalMsg) # compute the DFT/DFS of the modulating signal
SignalCar = f.fft(signalCar)  # compute the DFT/DFS of the carrier signal
SignalModulated = f.fft(signalModulated)  # compute the DFT/DFS of the modulated signal

normFrequ = np.arange(1,SignalMsg.size+1,dtype=float)/float(SignalMsg.size)
```

Let us plot the magnitude of each signal

```python
plt.subplot2grid((3,3), (0,0), colspan=3)
plot(normFrequ,abs(SignalMsg))
ylabel('|Message Signal|')
plt.subplot2grid((3,3), (1,0), colspan=3)
plot(normFrequ,abs(SignalCar))
ylabel('|Carrier Signal|')
plt.subplot2grid((3,3), (2,0), colspan=3)
plot(normFrequ,abs(SignalModulated))
xlabel('Samples')
ylabel('|AM Signal|')
show()
```

We would like to have a more precise insight on the peaks, plotting with the x-axis limits setted between 0.1 and 0.2,

```python
plt.subplot2grid((3,3), (0,0), colspan=3)
plot(normFrequ,abs(SignalMsg))
ylabel('|Message Signal|')
xlim([0,0.1])
plt.subplot2grid((3,3), (1,0), colspan=3)
plot(normFrequ,abs(SignalCar))
ylabel('|Carrier Signal|')
xlim([0.1,0.2])
plt.subplot2grid((3,3), (2,0), colspan=3)
plot(normFrequ,abs(SignalModulated))
xlabel('Samples')
xlim([0.1,0.2])
ylabel('|AM Signal|')
show()
```

Once again, the spectrum is translated around the carrier line
spectra.

Mapping into Symbols
--------------------

In AM transmissions, the signal is directly transmitted by
modulating the carrier.&nbsp; As seen in the lecture, many other
transmission approaches first map the values of the signal to be
transmitted into symbols, and them send such symbols. This is the
case of QAM, where the values of the signal are mapped into
symbols on the complex plane.

Let's generate the symbols of a QAM4

```python
M = 2 # Number of symbols = M^2
a = a = np.arange(-(M-1),M,2)

import cmath
comp = 1j

QAMsymbols = np.zeros((M,M),dtype = complex)

for i in range(M):
    for k in range(M):
        QAMsymbols[i,k] = a[i] + comp*a[k]

QAMsymbols = QAMsymbols.flatten()
```

We can then plot these symbols using the pylab.scatter(QAMSymbols.real,QAMSymbole.imag) function

```python
scatter(QAMsymbols.real,QAMsymbols.imag)
xlabel('Real Part')
ylabel('Imaginary Part')
for i in range(len(QAMsymbols)):
    text(QAMsymbols[i].real+0.1, QAMsymbols[i].imag+0.1, str(i))
show()
```

So consider a signal to be transmitted $x[n]$ that takes 4 possible values: 0, 1, 2, and 3.
We can simulate such signal by randomy generating integer values between 0 and 3. Suppose we want to generate 20 samples. In Python this can be done using the command

```python
N = 20  # number of symbols
x = np.random.randint(0,4,(N,))
```

The mapping from the signal to the symbols is then given by

```python
QAMsymbols_x = QAMsymbols[x]
```

This complex symbols sequence $S_k$ is first convolved with a (real) pulse $p(t)$ (e.g. rect function) of length $T_s$ to obtain a time-continuous complex signal
$$
S(t) = \sum_{k=0}^{N-1} S_k\, p(t - kT_s).
$$
In this example, we will take $p(t)$ to be the rectangular function of unit width.

To be transmitted, the signal still needs to be modulated by a carrier wave. This is done for QAM in the following way:
$$
s(t) = \Re \{S(t)\,e^{j2\pi f_0 t}\} = I(t)\cos(2\pi f_0 t) - Q(t)\sin(2\pi f_0 t),
$$
where $f_0$ is the carrier frequency, $I(t)=\Re\{S(t)\}$ and $Q(t)=\Im\{S(t)\}$.

It turns out both real (aka in-phase $I(t)$) and imaginary (aka quadrature $Q(t)$) parts of the signal can be recovered from $s(t)$. See [wikipedia](https://en.wikipedia.org/wiki/Quadrature_amplitude_modulation) for an explanation of the demodulation scheme.

Let's see now how the time-domain signals would look.

```python
Slen = 1  # Symbol length in periods of the carrier frequency
Fs = 16.  # We want to have 16 samples per period

n = np.arange(0, N*Slen*Fs)  # time index vector
S_index = np.array(np.floor(n/(Slen*Fs)), dtype=int)  # map to each time unit a symbol

I_x = QAMsymbols_x.real
Q_x = QAMsymbols_x.imag

# modulated signal
s = I_x[S_index]*np.cos(2*np.pi*n/Fs) - Q_x[S_index]*np.sin(2*np.pi*n/Fs)

plot(n/Fs, s)
title('QAM4 modulated signal')
xlabel('Time [1 symbol/unit time]')
ylabel('Signal amplitude')
show()
```
