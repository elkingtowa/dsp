### Symmetry of spectra and bandwidth limits

Python easily deals with samples indexed `0` on. Consequently, functions that are symmetric around the origin, like Fourier transforms of signals with real values, are rather represented with the symmetry around the middle index of the samples. To give an example,consider the prototypical band-limited function, that is, the sinc function. Its Fourier transform, as produced by the function `scipy.fftpack.fft()` will look like.

<img src="http://d396qusza40orc.cloudfront.net/dsp/num_examples/Module%206/images/Spectra.png">

rather than

<img src="https://d396qusza40orc.cloudfront.net/dsp/num_examples/Module%206/images/Spectra_2.png">

What we also have to point out is that the limits of the spectral support of the signal (bandwidth limits) are on the middle sample index (here 500 and 501) rather than in 1 and 1000. When expanding the spectral support of the signal (bandwidth) by zero padding, the zeros have to be placed at the bandwidth limits, therefore, at the middle sample index. This shall be clearer later on through examples.

### Normalized frequency

When talking about sampling, we are implicitly admitting that the samples come indeed from the sampling operation of a time continuous signals. Suppose that we have obtained $N$ samples, $x[1], ... , x[N]$ from sampling the signal $x(t)$.

Sample indices of the signal can be then linked to time via the sampling period $T_s$, where for the $k$-th sample of the signal we have $t=kT_s$. Similarly, sample indexes of a spectrum (DFT / DFS) can be related to frequencies via the sampling frequency $F_s=1/T_s$, where for the $k$-th sample of the spectrum we have $f=kF_s/N$.


When speaking of normalized frequency, the normalization is intended with respect to the sampling frequency $F_s$. The link between the continuous normalized frequency domain and the discrete one, is then $f/F_s=k/N$. Consequently, when expressed in terms of normalized frequencies, a DFT / DFS spectrum has a domain $1/N,...,(N-1)/N,1$.

## Sampling, Down-sampling, and Aliasing

### See the aliasing: A toy example

To give a visual idea of the aliasing problem we start with a toy example: A square sinc function. We can generate a square sinc function with the following code

```python
%pylab inline

import numpy as np
x = np.arange(-499,501,1,dtype=float)/20
indexZero = np.where(x == 0)
x[indexZero] = 1

from scipy import constants as c
y = pow(np.sin(c.pi*x)/(c.pi*x),2)
y[indexZero] = 1

from scipy import fftpack as f
Y = f.fft(y)

# Plot the DFT / DFS in normalized frequencies
N = len(y)
normFrequ = np.arange(1,N + 1,dtype = float)/float(N)

subplot(2,1,1)
plot(np.arange(N),y)
ylabel('y')
subplot(2,1,2)
plot(normFrequ,abs(Y))
xlabel('Normalized frequencies')
ylabel('|Y|')
show()
```

Clearly, we have a band-limited function. 

Let us now down sample our square sinc function $y$. 

We start by a factor of 5:

```python
sizey = len(y)
y5 = y[1:sizey:5]
Y5 = f.fft(y5)
```

obtaining

```python
N_5 = len(y5)

# Plot the DFT / DFS in normalized frequencies
normFrequY = np.arange(1,N_5+1,dtype=float)/float(N_5)

subplot(2,1,1)
plot(np.arange(N),y)
ylabel('y and y5')
plot(np.arange(N)[1:sizey:5],y5,'.')
subplot(2,1,2)
plot(normFrequY,abs(Y5))
xlabel('Normalized frequencies')
ylabel('|Y5|')
show()
```

The dots represent `y5`, and the blue curve represents `y`.

The signal `y5` is still band-limited and we can recognize the spectrum of the original signal `y`.


We now down-sample by a factor 10:

```python
sizey = len(y)
y10 = y[1:sizey:10]
Y10 = f.fft(y10)
```

obtaining

```python
N_10 = len(y10)
    
# Plot the DFT / DFS in normalized frequencies
normFrequy_10 = np.arange(1,N_10+1,dtype=float)/float(N_10)

subplot(2,1,1)
plot(np.arange(N),y)
ylabel('y and y10')
plot(np.arange(N)[1:sizey:10],y10,'.')
subplot(2,1,2)
plot(normFrequy_10,abs(Y10))
xlabel('Normalized frequencies')
ylabel('|Y10|')
show()
```

We are now at the borderline of the band-limit, yet the spectrum of the original signal can still be recognized.

We proceed by down-sampling of a factor 20

```python
sizey = len(y)
y20 = y[1:sizey:20]
Y20 = f.fft(y20)
```

obtaining

```python
N_20 = len(y20)

# Plot the DFT / DFS of the down-sampled signal in normalized frequencies
normFrequy_20 = np.arange(1,N_20+1,dtype=float)/float(N_20)

subplot(2,1,1)
plot(np.arange(N),y)
ylabel('y and y20')
plot(np.arange(N)[1:sizey:20],y20,'.')
subplot(2,1,2)
plot(normFrequy_20,abs(Y20))
xlabel('Normalized frequencies')
ylabel('|Y20|')
show()
```

The spectrum of the down-sampled signal is now affected by aliasing and is a distorted version of the original spectrum, the latter not being anymore recognizable. 

The original 1000 sample signal can be reconstructed by zero padding the spectrum (bandwidth expansion) and performing an inverse DFT / DFS. As mentioned above, zero padding is performed at the middle index of the spectrum. Let's compare the reconstructed form of the three down-sampled signals

```python
sizeY = len(Y)

N_Y5 = len(Y5)

# Zero Padding
YY5 = np.append(Y5[0:N_Y5/2],np.zeros(sizeY - N_Y5))
YY5 = np.append(YY5,Y5[N_Y5/2:N_Y5])

# Scaling of the DFT / DFS due to the increased number of samples
scailingFactor =float(sizeY)/float(N_Y5)

# Scaled inverse DFT / DFS. 
# Notice that the signal is must be real and any imaginary part is due to numerical errors
yy5 = scailingFactor * np.real(f.ifft(YY5))

N_Y10 = len(Y10)

# Zero Padding
YY10 = np.append(Y10[0:N_Y10/2],np.zeros(sizeY - N_Y10))
YY10 = np.append(YY10,Y10[N_Y10/2:N_Y10])

# Scaling of the DFT / DFS due to the increased number of samples
scailingFactor =float(sizeY)/float(N_Y10)

# Scaled inverse DFT / DFS. 
# Notice that the signal is must be real and any imaginary part is due to numerical errors
yy10 = scailingFactor * np.real(f.ifft(YY10))

N_Y20 = len(Y20)

# Zero Padding
YY20 = np.append(Y20[0:N_Y20/2],np.zeros(sizeY - N_Y20))
YY20 = np.append(YY20,Y20[N_Y20/2:N_Y20])

# Scaling of the DFT / DFS due to the increased number of samples
scailingFactor =float(sizeY)/float(N_Y20)

# Scaled inverse DFT / DFS. 
# Notice that the signal is must be real and any imaginary part is due to numerical errors
yy20 = scailingFactor * np.real(f.ifft(YY20))

subplot2grid((3,3), (0, 0), colspan=2)
plot(np.arange(yy5.size),yy5)
plot(np.arange(N),y,'-')
ylabel('yy5 and y')
subplot2grid((3,3), (1, 0), colspan=2)
plot(np.arange(yy10.size),yy10)
plot(np.arange(N),y,'-')
ylabel('yy10 and y')
subplot2grid((3,3), (2, 0), colspan=2)
plot(np.arange(yy20.size),yy20)
plot(np.arange(N),y,'-')
ylabel('yy20 and y')
xlabel('Samples')
show()
```

In the first two cases we have a prefect reconstruction of the 1000 sample signal $y$, while in the third case we obtain a different signal (a sort of sinc): The spectrum is aliased and the original information is lost.

### Hear the aliasing: Three tones again!

After our visual aliasing experience we move on to an auditory one. Recall the three tones we met in Module 4:

```python
import numpy as np
from scipy import constants as c
y = np.linspace(0,999,1000)
x1 = np.sin((y*2*c.pi*40)/1000)
x2 = np.sin((y*2*c.pi*80)/1000)
x3 = np.sin((y*2*c.pi*160)/1000)
x = np.append(x1,[np.zeros(1000),x2,np.zeros(1000),x3])
```

You can play it via audacity or winsound. We will use here a sampling frequency of 9000 Hz.

We recall that the spectrum of the signal x is:

```python
N = len(x)
X = f.fft(x)

# Plot the DFT / DFS of the down-sampled signal in normalized frequencies
normFrequ = np.arange(1,N+1,dtype = float)/float(N)

plot(normFrequ,abs(X))
xlabel('Normalized Frequencies')
ylabel('|X|')
show()
```

Let us down-sample the signal by a factor of 2 and plot the spectrum

```python
sizex = len(x)

# Down-sampling by a factor of 2
x2 = x[1:sizex:2]
N_2 = len(x2)

# DFT / DFS of the down-sampled signal
X2 = f.fft(x2)

# Plot of the spectrum of the down-sampled signal in normalized frequencies
normFrequN_2 = np.arange(1,N_2+1,dtype = float)/float(N_2)

plot(normFrequN_2,abs(X2))
xlabel('Normalized Frequencies')
ylabel('|2|')
show()
```

and then reconstruct the 5000 samples signal by zero padding the spectrum

```python
# Zero padding
XX2 = np.append(X2[0:N_2/2],np.zeros(sizex - N_2))
XX2 = np.append(XX2,X2[N_2/2:N_2])
# Reconstruction by inverse DFT /DFS
xx2 = np.real(f.ifft(XX2))
```

Play `xx2` and you will hear no difference with the original signal `x`.

Now it is time to down-sample by a factor of 10.

```python
sizex = len(x)

# Down-sampling by a factor of 10$x2 = x[1:sizex:2]
x10 = x[1:sizex:10]
N_10 = len(x10)

# DFT / DFS of the down-sampled signal
X10 = f.fft(x10)

# Plot of the spectrum of the down-sampled signal in normalized frequencies
normFrequN_10 = np.arange(1,N_10+1,dtype = float)/float(N_10)

plot(normFrequN_10,abs(X10))
xlabel('Normalized Frequencies')
ylabel('|X10|')
show()
```

The spectral content of the down-sampled signal `x10` is now aliased and it is quite different from the spectral content of the original signal `x`. We can therefore imagine that reconstructing the original 5000 samples signal by zero padding the spectrum will fail and rather provide a different signal. Let's check this out:

```python

# Zero padding
XX10 = np.append(X10[1:N_10/2],np.zeros(sizex - N_10))
XX10 = np.append(XX10,X10[N_10/2:N_10])

# Reconstruction by inverse DFT /DFS
xx10 = np.real(f.ifft(XX10))
```

Now play this one! ... weird isn't it? The original information has been distorted by the aliasing effect! and indeed it is not possible to recognize.

Once again, do a paper and pencil verification of what has happened!! (hint: you can consider the original spectrum $X$ to be zero in the normalized frequency period [0.2,0.8]).

You should remark that when down-sampling by a factor of 10, the spectral content of the second and third tone are aliased, corrupting the whole spectral content.

## Down-sampling and the sampling theorem: Better lose the saddle than the horse

### Jingle!

Load the Christmas jingle with a Swiss yodeling touch that we have introduced in Module 5 ([jingle.mat](https://d396qusza40orc.cloudfront.net/dsp/num_examples/Module%206/code/jingle.mat)). The files contains 2 variables: `jingle`, which is the vector containing the samples of the sound signal of the jingle, and `Fs`, which is the sampling frequency (frequency at which the continuous time sound signal has been sampled).

You can plot the DFT / DFS to take a look at the spectral content of the signal

```python
# Loading the jingle
import scipy.io as sio
mat_contents = sio.loadmat('jingle.mat')
jingle = mat_contents['jingle']
jingle = jingle.reshape(jingle.size,)

N_J = len(jingle)

# DFT / DFS of the jingle
Jingle = f.fft(jingle)

JingleLP = Jingle

Jingle = f.fft(jingle)

# Plot of the spectrum in normalized frequencies
NormFrequJ = np.arange(1,N_J + 1,dtype = float)/float(N_J)

plot(NormFrequJ,abs(Jingle))
xlabel('Normalized Frequencies')
ylabel('|Jingle|')
show()
```

The plot suggests that the spectral contents are located between 0 and approximately 0.24 (and symmetrically between 0.76 and 1).
So there should be no problem in down-sampling the signal by a factor of 2. Let's try to do so and then to reconstruct the full signal by zero padding the Fourier transform.

```python
from scipy import fftpack as f
import numpy as np

sizejingle = len(jingle)
#Down-sampling by a factor of 2
jingle2 = jingle[1:sizejingle:2]
N_J2 = len(jingle2)
# DFT / DFS of the down-sampled jingle
Jingle2 = f.fft(jingle2)
# Zero Padding
JJingle2 = np.append(Jingle2[0:N_J2/2],np.zeros(sizejingle - N_J2))
JJingle2 = np.append(JJingle2,Jingle2[N_J2/2 : N_J]) * float(sizejingle)/float(N_J2)
# Reconstruction by inverse DFT /DFS
jjingle2 = np.real(f.ifft(JJingle2))
```

Now play the original signal `jingle` and the signal `jjingle2` reconstructed from the down-sampled signal `jingle2`: You will hear no substantial difference! Using a 44k sampling frequency yields a nice sound.

Let us know down-sample the original signal by a factor of 10. Now we expect an aliasing effect and therefore that the reconstructed signal will be distorted. Let's hear that!

```python
sizejingle = len(jingle)
#Down-sampling by a factor of 2
jingle10 = jingle[1:sizejingle:10]
N_J10 = len(jingle10)
# DFT / DFS of the down-sampled jingle
Jingle10 = f.fft(jingle10)
# Zero Padding
JJingle10 = np.append(Jingle10[0:N_J10/2],np.zeros(sizejingle - N_J10))
JJingle10 = np.append(JJingle10,Jingle10[N_J10/2 : N_J]) *float(sizejingle)/float(N_J10)
# Reconstruction by inverse DFT /DFS
jjingle10 = np.real(f.ifft(JJingle10))
```

If we now play the signal we can remark that it has been severely distorted by the aliasing effect and the information contained in the original signal has been altered! In summary, we have lost the information!

If we really want to stick with our idea of down-sampling the jingle by a factor of 10, we can reduce the initial information so that it is not distorted by the aliasing effect. That is, we diligently follow the sampling theorem and low-pass the signal before down-sampling it. It is better to lose part of the information than having the whole spectral content corrupted by aliasing.


Let's see how ...
We start by low-passing the jingle. Given that we aim a down-sampling by a factor of 10, we have to cut all the spectral content from $0.05$ to $0.5$ and, symmetrically, from $0.5$ to $0.95$. We do so in the frequency domain:

```python
N_J1 = len(Jingle)
N1 = N_J1 * 0.05
N2 = N_J1 * 0.95

# Low pass filtering of the jingle spectrum
JingleLP[N1:N2+1] = np.zeros(N2-N1+1)

# Lowpass version of the jingle
jingleLP = np.real(f.ifft(JingleLP))
```

Play $jingleLP$ to discover that the jingle sounds less loud (low passed) but it is not distorted!

Let us take a look at the spectral content, so to remark the low passing effect


```python
# Plot in normalized Frequencies
normFrequN_J = np.arange(1,N_J1+1,dtype = float)/float(N_J1)

subplot(2,1,1)
plot(normFrequN_J,abs(Jingle))
ylabel('|Jingle|')
subplot(2,1,2)
plot(normFrequN_J,abs(JingleLP))
xlabel('Normalized Frequencies')
ylabel('|JingleLP|')
show()
```

Now down-sample the low passed signal and reconstruct the original low passed signal from the down-sampled data:

```python
sizejingle = len(jingleLP)
# Down-sample the low pass version of the jingle by a factor 10
jingleLP10 = jingleLP[1:N_J1:10]
N_J10LP = len(jingleLP10)
# DFT / DFS of the down-sampled low pass version of the jingle
JingleLP10 = f.fft(jingleLP10)
# Zero padding
JJingleLP10 = np.append(JingleLP10[1:N_J10LP/2],np.zeros(sizejingle - N_J10LP))
JJingleLP10 = np.append(JJingleLP10,JingleLP[N_J10LP/2:N_J10LP]) * float(sizejingle)/float(N_J10LP)
# Reconstruction by inverse DFT /DFS
jjingleLP10 = np.real(f.ifft(JJingleLP10))
```

You can play `jjingleLP10` and `jjingle10` to remark that the first sounds less loud but it is not distorted, while the second it is indeed distorted!

So, we'd better stick to the sampling theorem and lose the saddle but not the horse!!!
